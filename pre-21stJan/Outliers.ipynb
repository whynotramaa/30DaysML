```python
p25 = data['CGPA'].quantile(0.25)
p75 = data['CGPA'].quantile(0.75)
```

```python
hw = hw.rename(columns={'Height(Inches)':'ht'})
hw = hw.rename(columns={'Weight(Pounds)':'wt'})
```

```python
hw.describe()
```

```python
h_skew = hw['ht'].skew()
w_skew = hw['wt'].skew()
```

```python
print(hw['ht'].min(), hw['ht'].max())
```

```python
h_p25 = hw['ht'].quantile(0.25)
h_p75 = hw['ht'].quantile(0.75)

w_p25 = hw['wt'].quantile(0.25)
w_p75 = hw['wt'].quantile(0.75)

print(h_p25)
print(h_p75)
print(w_p25)
print(w_p75)
```

```python
h_iqr = h_p75 - h_p25
w_iqr = w_p75 - w_p25

h_lower_limit = h_p25 - 1.5*h_iqr
h_upper_limit = h_p75 + 1.5*h_iqr

w_lower_limit = w_p25 - 1.5*w_iqr
w_upper_limit = w_p75 + 1.5*w_iqr

print(h_lower_limit)
print(h_upper_limit)
print(w_lower_limit)
print(w_upper_limit)
```

```python
h_outliers = hw[ (hw['ht'] < h_lower_limit) | (hw['ht'] > h_upper_limit) ]
w_outliers = hw[ (hw['wt'] < w_lower_limit) | (hw['wt'] > w_upper_limit) ]
# or doesnt work in here we must do |
print(h_outliers)
print(w_outliers)
```

```python
hw = hw[ (hw['ht'] >= h_lower_limit) & (hw['ht'] <= h_upper_limit) ]
hw = hw[ (hw['wt'] >= w_lower_limit) & (hw['wt'] <= w_upper_limit) ]
```
